```python
from lcapy import n, delta, Omega, z, k
import numpy as np
import matplotlib.pyplot as plt
```

```python
def plot_transforms(h, xlim=(-10, 10)):

    # plot impulse response
    h.plot(xlim, figsize=(6, 2)) #, ylim = [-1, 1])

    # DTFT
    H = h.DTFT(Omega)
    H.plot(
        Wvector=(-np.pi, np.pi), 
        plot_type='mag-phase', 
        **{
            'xscale': 1/np.pi,
        },
    )

    # Z-transform
    Z = h(z)
    Z.plot()

    # DFT
    K = h(k)
    K.plot(xlim, figsize=(6, 2))

    return H, Z, K
```

```python
h = delta(n)
H, Z, K = plot_transforms(h)
H
```

```python
h = delta(n+1)
H, Z, K = plot_transforms(h)
H
```

```python
h = 0.5*(delta(n) + delta(n + 1))
H, Z, K = plot_transforms(h)
H
```
